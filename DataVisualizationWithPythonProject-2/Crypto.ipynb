## CODING TASK #4. PLOT HISTOGRAMS USING MATPLOTLIB
- **A histogram represents data using bars with various heights**
- **Each bar groups numbers into specific ranges**
- **Taller bars show that more data falls within that specific range**
- **Plot the histogram of the returns**

```python
mu = round(daily_return_df['BTC'].mean(), 2) # mean of distribution
sigma = round(daily_return_df['BTC'].std(), 2) # standard deviation of distribution
 
num_bins = 40
print(mu)
print(sigma)
```

```python
# Plot the histogram of the returns

plt.figure(figsize = (10, 6))
plt.hist(daily_return_df['BTC'], num_bins, facecolor = 'blue');
plt.grid()

plt.title('BTC: mu = ' + str(mu) + ', sigma: ' + str(sigma))
```

```python

print(mu)
print(sigma)
```

```python
# Plot the histogram of the returns
mu1 = round(daily_return_df['ETH'].mean(), 2) # mean of distribution
sigma1 = round(daily_return_df['ETH'].std(), 2) # standard deviation of distribution

num_bins = 40

plt.figure(figsize = (10, 6))
plt.hist(daily_return_df['ETH'], num_bins, facecolor = 'blue');
plt.grid()

plt.title('ETH: mu = ' + str(mu1) + ', sigma: ' + str(sigma1))
```

```python
# Plot the histogram of the returns
mu2 = round(daily_return_df['ADA'].mean(), 2) # mean of distribution
sigma2 = round(daily_return_df['ADA'].std(), 2) # standard deviation of distribution

num_bins = 40

plt.figure(figsize = (10, 6))
plt.hist(daily_return_df['ADA'], num_bins, facecolor = 'blue');
plt.grid()

plt.title('ADA: mu = ' + str(mu2) + ', sigma: ' + str(sigma2))
```

**PRACTICE OPPORTUNITY #4:**
- **Plot the histogram for BTC and ETH returns using 60 bins. Display all histograms on the same figure**


```python
# Plot the histogram for BTC and ETH returns using 60 bins. Display all histograms on the same figure

 
num_bins = 60

fig, ax = plt.subplots(figsize=(10, 7))
ax.hist(daily_return_df['BTC'], num_bins, density=True, histtype='bar', rwidth=0.8, color='blue', label = 'BTC')
ax.hist(daily_return_df['ETH'], num_bins, density=True, histtype='bar', rwidth=0.8, color='yellowgreen', label = 'ETH')

ax.set_title('BTC and ETH returns')
ax.set_xlabel('Daily Returns')
ax.set_ylabel('Frequency')
plt.grid()
plt.show()

```

```python
# Plot scatter plot between mean area and mean smoothness

plt.figure(figsize = (10, 10))
sns.scatterplot(data=cancer_df, x="mean area", y="mean smoothness", 
                hue= 'target', 
                size='mean area',
                sizes=(20, 300), 
                alpha = 0.8,
                palette = "plasma")
plt.title('Mean area and Mean smoothness');
```

```python
cancer_df.target.value_counts()
```

```python
# Let's print out countplot to know how many samples belong to class #0 and #1

fig, ax = plt.subplots(figsize=(6,7))

ax=sns.countplot(x=cancer_df['target'], data=cancer_df, palette = "Accent")

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+ 0.3));
    
```

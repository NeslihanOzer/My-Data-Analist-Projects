```python
# Using Matplotlib, plot lineplots that display all 4 stocks daily prices on one single figure.

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(stocks_daily_prices['Date'], stocks_daily_prices['AAPL'], label='Apple', linewidth = 1)
ax.plot(stocks_daily_prices['Date'], stocks_daily_prices['JPM'], label='J.P. Morgan', linewidth = 1)
ax.plot(stocks_daily_prices['Date'], stocks_daily_prices['PG'], label='Procter & Gamble', linewidth = 1)
ax.plot(stocks_daily_prices['Date'], stocks_daily_prices['UAL'], label='United Airlines Holdings', linewidth = 1)

ax.set_xlabel('Date')
ax.set_ylabel('Price')
ax.set_title('Stock Prices')

ax.legend()
plt.grid(True)
plt.show()
```

```python
# Using Matplotlib, plot 4 stocks daily prices on multiple subplots.

fig, ax = plt.subplots(4, 1, figsize=(15, 10))

ax[0].plot(stocks_daily_prices['Date'], stocks_daily_prices['AAPL'], linewidth = 2)
ax[0].set_ylabel('Price [$]')
ax[0].set_title('AAPL Stock Prices')
ax[0].grid(True)

ax[1].plot(stocks_daily_prices['Date'], stocks_daily_prices['JPM'], linewidth = 2, color='r')
ax[1].set_ylabel('Price [$]')
ax[1].set_title('JPM Stock Prices')
ax[1].grid(True)

ax[2].plot(stocks_daily_prices['Date'], stocks_daily_prices['PG'], linewidth = 2, color='g')
ax[2].set_ylabel('Price [$] ')
ax[2].set_title('PG Stock Price')
ax[2].grid(True)

ax[3].plot(stocks_daily_prices['Date'], stocks_daily_prices['UAL'], linewidth = 2, color='orange')
ax[3].set_ylabel('Price [$] ')
ax[3].set_title('UAL Stock Price')
ax[3].grid(True)

plt.xlabel('Date')
plt.tight_layout()
plt.show()

# Apple hisse senedi fiyatlari 2020 yilindan sonra ivmelenerek artmis olup 2022 yilindan sonra dususe baslamistir.
# JP Morgan hisse sendi fiyatlari 2017-2022 yillari arasinda artmis olup 2022 yilindan sonra dususe baslamistir. 
# Procter & Gamble in hisse senedi fiyatlari JP Morgan ile benzerlik gostermektedir.
# United Airlines Holdings in hisse senedi fiyatlari 2010 yilindan itibaren dalgali bir sekilde yukselmis ancak
# 2020 yilinda sert bir dusus gerceklestirmistir
```

```python
# Using Matplotlib, plot the 4 plots on subplots next to each other (all figures in one row)

fig, ax = plt.subplots(1, 4, figsize=(12, 4))

ax[0].plot(stocks_daily_prices['Date'], stocks_daily_prices['AAPL'], linewidth = 1)
ax[0].set_ylabel('Price [$]')
ax[0].set_title('AAPL Stock Prices')
ax[0].grid(True)

ax[1].plot(stocks_daily_prices['Date'], stocks_daily_prices['JPM'], linewidth = 1, color='r')
ax[1].set_ylabel('Price [$]')
ax[1].set_title('JPM Stock Prices')
ax[1].grid(True)


ax[2].plot(stocks_daily_prices['Date'], stocks_daily_prices['PG'], linewidth = 1, color='g')
ax[2].set_ylabel('Price [$] ')
ax[2].set_title('PG Stock Price')
ax[2].grid(True)


ax[3].plot(stocks_daily_prices['Date'], stocks_daily_prices['UAL'], linewidth = 1, color='orange')
ax[3].set_ylabel('Price [$] ')
ax[3].set_title('UAL Stock Price')
ax[3].grid(True)


plt.xlabel('Date')
plt.tight_layout()


plt.show()
```

```python
# Using Seaborn, plot similar scatterplot between Apple and JP Morgan daily returns.

plt.figure(figsize=(10, 6))
sns.scatterplot(data=stocks_daily_returns, x=stocks_daily_returns['AAPL'], y=stocks_daily_returns['JPM'])

plt.xlabel('Apple Daily Returns')
plt.ylabel('JP Morgan Daily Returns')
plt.title('Apple vs JP Morgan Daily Returns')
plt.grid()
plt.show()
```

```python
# Assume that you decided to become bullish on AAPL and you allocated 70% of your assets in it. 
#You also decided to equally divide the rest of your assets in other stocks (JPM, PG, and UAL). 
#Using Matplotlib, plot a pie chart that shows these allocations. 
#Use 'explode’ attribute to increase the separation between AAPL and the rest of the portfolio.


labels = ["Apple", "JPM", "PG", "UAL"]
sizes = [70, 10, 10, 10]
colors = ['g', 'r', 'y', 'b']
explode = [0.1, 0, 0, 0]


plt.figure(figsize = (5, 5))
plt.pie(sizes, colors=colors, labels = labels, explode = explode, 
        shadow=True, autopct='%1.1f%%', startangle=90, hatch=['/', 0, 0, 0, 0],)

plt.title('Stock Portfolio')
```

```python
# Using Matplotlib, plot the histogram for United Airlines and P&G returns using 40 bins with red color. 
# Display the mean and Standard deviation for both stocks on top of the figure. 
# What did you conclude from this graph?

mu = stocks_daily_returns['UAL'].mean()
sigma = stocks_daily_returns['PG'].std()

num_bins = 40

fig, ax = plt.subplots(figsize=(8, 6))
ax.hist(stocks_daily_returns['UAL'], num_bins, density=True, histtype='bar', rwidth=0.8, color='blue')
ax.hist(stocks_daily_returns['PG'], num_bins, density=True, histtype='bar', rwidth=0.8, color='yellowgreen')

ax.set_title('United Airlines and P&G returns')
ax.set_xlabel('Returns')
ax.set_ylabel('Frequency')
plt.grid()
plt.show()


```

```python
# Using Seaborn, plot a heatmap that shows the correlations between stocks daily returns. 

sns.heatmap(stocks_daily_returns.corr(), annot=True, cmap='viridis')

plt.title('Correlation Heatmap of Stocks Daily Returns')

plt.show()
```

```python
# Comment on the correlation between UAL and P&G.

correlation = stocks_daily_returns['UAL'].corr(stocks_daily_returns['PG'])

print("Correlation between UAL and P&G: {:.2f}".format(correlation))

```

```python
# Plot a 3D plot showing all daily returns from JPM, AAPL and UAL [External Research is required].


from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(111, projection = '3d')

ax.scatter(stocks_daily_returns['AAPL'], 
           stocks_daily_returns['JPM'], 
           stocks_daily_returns['UAL'],
            c = 'b', s = 100) # c for color, s for size of each points
ax.set_xlabel('AAPL')
ax.set_ylabel('JPM')
ax.set_zlabel('UAL')
ax.set_title('Stock daily returns')
```

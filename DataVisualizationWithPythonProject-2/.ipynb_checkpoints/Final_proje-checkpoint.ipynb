```python
# Using Matplotlib, plot lineplots that display all 4 stocks daily prices on one single figure.


fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(stocks_daily_prices['Date'], stocks_daily_prices['AAPL'], label='Apple', linewidth = 1, color='blue')
ax.plot(stocks_daily_prices['Date'], stocks_daily_prices['JPM'], label='J.P. Morgan', linewidth = 1, color='red')
ax.plot(stocks_daily_prices['Date'], stocks_daily_prices['PG'], label='Procter & Gamble', linewidth = 1, color='lime')
ax.plot(stocks_daily_prices['Date'], stocks_daily_prices['UAL'], label='United Airlines Holdings', linewidth = 1, color='orange')

ax.set_xlabel('Date')
ax.set_ylabel('Price')
ax.set_title('Stock Prices')

ax.legend()
plt.grid()
plt.show()
```

```python
# Using Matplotlib, plot 4 stocks daily prices on multiple subplots.

fig, ax = plt.subplots(4, 1, figsize=(15, 10))

ax[0].plot(stocks_daily_prices['Date'], stocks_daily_prices['AAPL'], linewidth = 1, color='blue')
ax[0].set_ylabel('Price [$]')
ax[0].set_title('AAPL Stock Prices')
ax[0].fill_between(stocks_daily_prices['Date'], 0, stocks_daily_prices['AAPL'], alpha=0.1, color='blue')
ax[0].grid(True)

ax[1].plot(stocks_daily_prices['Date'], stocks_daily_prices['JPM'], linewidth = 1, color='red')
ax[1].set_ylabel('Price [$]')
ax[1].set_title('JPM Stock Prices')
ax[1].fill_between(stocks_daily_prices['Date'], 0, stocks_daily_prices['JPM'], alpha=0.1, color='red')
ax[1].grid(True)

ax[2].plot(stocks_daily_prices['Date'], stocks_daily_prices['PG'], linewidth = 1, color='lime')
ax[2].set_ylabel('Price [$] ')
ax[2].set_title('PG Stock Price')
ax[2].fill_between(stocks_daily_prices['Date'], 0, stocks_daily_prices['PG'], alpha=0.1, color='lime')
ax[2].grid(True)

ax[3].plot(stocks_daily_prices['Date'], stocks_daily_prices['UAL'], linewidth = 1, color='orange')
ax[3].set_ylabel('Price [$] ')
ax[3].set_title('UAL Stock Price')
ax[3].fill_between(stocks_daily_prices['Date'], 0, stocks_daily_prices['UAL'], alpha=0.1, color='orange')
ax[3].grid(True)

plt.xlabel('Date')
plt.tight_layout()
plt.show()

# Apple stock prices accelerated after 2020 and started to decline after 2022.
# From the first quarter of 2020, the negative effects of the pandemic have started to be seen all over the world. 
# JP Morgan and United Airlines Holdings were affected more by the pandemic than the other three stocks. 
# Air transport company UHL has been the most affected company due to travel restrictions.
# Procter & Gamble has been relatively less affected by the pandemic due to the type of products it sells.
```

```python
# Using Matplotlib, plot the 4 plots on subplots next to each other (all figures in one row)

fig, ax = plt.subplots(1, 4, figsize=(20, 10))

ax[0].plot(stocks_daily_prices['Date'], stocks_daily_prices['AAPL'], linewidth = 1, color='blue')
ax[0].set_ylabel('Price [$]')
ax[0].set_title('AAPL Stock Prices')
ax[0].grid(True)

ax[1].plot(stocks_daily_prices['Date'], stocks_daily_prices['JPM'], linewidth = 1, color='red')
ax[1].set_ylabel('Price [$]')
ax[1].set_title('JPM Stock Prices')
ax[1].grid(True)

ax[2].plot(stocks_daily_prices['Date'], stocks_daily_prices['PG'], linewidth = 1, color='lime')
ax[2].set_ylabel('Price [$] ')
ax[2].set_title('PG Stock Price')
ax[2].grid(True)

ax[3].plot(stocks_daily_prices['Date'], stocks_daily_prices['UAL'], linewidth = 1, color='orange')
ax[3].set_ylabel('Price [$] ')
ax[3].set_title('UAL Stock Price')
ax[3].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Using Seaborn, plot similar scatterplot between Apple and JP Morgan daily returns.

plt.figure(figsize=(10, 6))
sns.scatterplot(data=stocks_daily_returns, x=stocks_daily_returns['AAPL'], y=stocks_daily_returns['JPM'])

plt.xlabel('Apple Daily Returns')
plt.ylabel('JP Morgan Daily Returns')
plt.title('Apple vs JP Morgan Daily Returns')
plt.grid()
plt.show()
```

```python
# Assume that you decided to become bullish on AAPL and you allocated 70% of your assets in it. 
#You also decided to equally divide the rest of your assets in other stocks (JPM, PG, and UAL). 
#Using Matplotlib, plot a pie chart that shows these allocations. 
#Use 'explode’ attribute to increase the separation between AAPL and the rest of the portfolio.


labels = ["Apple", "JPM", "PG", "UAL"]
sizes = [70, 10, 10, 10]
colors = ['blue', 'red', 'green', 'yellow']
explode = [0.1, 0, 0, 0]


plt.figure(figsize = (5, 5))
plt.pie(sizes, colors=colors, labels = labels, explode = explode, 
        shadow=True, autopct='%1.1f%%', startangle=90)

plt.title('Stock Portfolio');
```

```python
# Using Matplotlib, plot the histogram for United Airlines and P&G returns using 40 bins with red color. 
# Display the mean and Standard deviation for both stocks on top of the figure. 
# What did you conclude from this graph?

mu = "{:.2f}".format(stocks_daily_returns['UAL'].mean())
sigma = "{:.2f}".format(stocks_daily_returns['UAL'].std())

mu1 = "{:.2f}".format(stocks_daily_returns['PG'].mean())
sigma1= "{:.2f}".format(stocks_daily_returns['PG'].std())

num_bins = 40

fig, ax = plt.subplots(1, 2, figsize=(15, 5))

ax[0].hist(stocks_daily_returns['UAL'], num_bins, density=True, histtype='bar', rwidth=0.8, color='red', label='UAL')
ax[0].set_ylabel('Frequency')
ax[0].set_title('United Airlines: mu = ' + str(mu) + ', sigma: ' + str(sigma))
ax[0].grid()
ax[0].legend()

ax[1].hist(stocks_daily_returns['PG'], num_bins, density=True, histtype='bar', rwidth=0.8, color='red', label='PG')
ax[1].set_title('P&G: mu = ' + str(mu1) + ', sigma: ' + str(sigma1))
ax[1].grid()
ax[1].legend()


plt.show()



# Both stocks have a normal distribution

# Standard deviation is often used as a measure of risk or volatility in finance and statistics. 
# A larger standard deviation implies higher risk due to the wider range of potential outcomes and greater variability in the dataset. 
# In contrast, a smaller standard deviation indicates lower risk and reduced uncertainty since the data points are more tightly clustered around the mean.

# P&G has lower risk and reduced uncertainty than UAL

```

```python
# Using Seaborn, plot a heatmap that shows the correlations between stocks daily returns. 

sns.heatmap(stocks_daily_returns.corr(), annot=True, cmap='plasma')

plt.title('Correlation Heatmap of Stocks Daily Returns')

plt.show()
```

```python
# Comment on the correlation between UAL and P&G.

correlation = stocks_daily_returns['UAL'].corr(stocks_daily_returns['PG'])

print("Correlation between UAL and P&G: {:.2f}".format(correlation))

```

```python
# Plot a 3D plot showing all daily returns from JPM, AAPL and UAL [External Research is required].


from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(111, projection = '3d')

ax.scatter(stocks_daily_returns['AAPL'], 
           stocks_daily_returns['JPM'], 
           stocks_daily_returns['UAL'],
            c = 'b', s = 100) # c for color, s for size of each points
ax.set_xlabel('AAPL')
ax.set_ylabel('JPM')
ax.set_zlabel('UAL')
ax.set_title('Stock daily returns')
```
